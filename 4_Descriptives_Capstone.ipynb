```python
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import numpy as np
import tabulate
from pandas import ExcelWriter
```

#Upload Data

```python
data_path = '/content/descriptives (2).xlsx'
df =  pd.read_excel(data_path, index_col=False)
```

#Descriptives
Percentage of different attributes

```python
varibales = ["Age", "Nationality", "Income", "Occupation", "Gender", "Education"]
w = ExcelWriter('desFreq.xlsx')
for x in varibales:
  table = df[x].value_counts()
  keys = table.keys()
  Vals = table.values
  freq = pd.DataFrame({x:keys, 'Frequency' : Vals})
  freq['Percent']=freq['Frequency']/freq['Frequency'].sum()
  freq.to_excel(w, sheet_name=str(x), index=False)
  w.save()
    
```
